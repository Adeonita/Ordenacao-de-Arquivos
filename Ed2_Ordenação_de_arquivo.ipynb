#Implementação dos Agoritmos de Ordenação


A biblioteca Pandas já dispõe de um método para realizar a ordenação dos seus Dataframes, o **sort.values( )** que recebe 3 parâmetros, **by**, **ascending** e **kind**. Campo, ordem e tipo, respectivamente.


Lista para armazenar todos os tempos de execução

```
tempos_de_execucao = []
```

#TimeIt

Para mensurar o tempo de execução é necessário envolver o método sort_values( ) da bibioteca panda no método timeit.default_timer( ) do módulo timeit, determinando o início e o fim da função, para que através da diferença entre eles seja calculado o tempo de execução da função de ordenação.

#Quick Sort



###Ordem: Ascendente 
####Campo: Acession
---


```
inicio = timeit.default_timer()
crescente_acession_quickSort = arquivo.sort_values( by='Acession', ascending=True, kind='quicksort')
fim = timeit.default_timer()
tempo_execucao_acendente_acession_quick = fim - inicio
tempos_de_execucao.append(['ASC ACESSION quicksort:', tempo_execucao_acendente_acession_quick])
print("\n O tempo de execução da função quickSort ascendente no campo ACESSION foi de:{} \n".format(tempo_execucao_acendente_acession_quick))
crescente_acession_quickSort
```

###Ordem: Descendente 
####Campo: Acession
---

```
inicio = timeit.default_timer()
decrescente_acession_quickSort = arquivo.sort_values( by='Acession', ascending=False, kind='quicksort')
fim = timeit.default_timer()
tempo_execucao_descendente_acession_quick = fim - inicio
tempos_de_execucao.append(['DESC ACESSION quicksort:', tempo_execucao_descendente_acession_quick])
print("\nO tempo de execução da função quickSort descendente no campo ACESSION foi de: {} \n".format(tempo_execucao_descendente_acession_quick))
decrescente_acession_quickSort
```

###Ordem: Ascendente 
####Campo: Date
---

```
inicio = timeit.default_timer()
crescente_date_quickSort = arquivo.sort_values( by='Date', ascending=True, kind='quicksort')
fim = timeit.default_timer()
tempo_execucao_ascendente_date_quick = (fim - inicio)
tempos_de_execucao.append(['ASC DATE quicksort:', tempo_execucao_ascendente_date_quick])
print("\nO tempo de execução da função quickSort ascendente no campo DATE foi de: {} \n".format(tempo_execucao_ascendente_date_quick))
crescente_date_quickSort
```

```
tempos_de_execucao 

%matplotlib inline

tempos_legenda = [];
tempos = [];
for line in tempos_de_execucao:
  tempos_legenda.append(line[0])
  tempos.append(line[1])

import matplotlib.pyplot as plt
plt.rcParams['figure.figsize'] = [22, 4]
y_pos = np.arange(len(tempos_legenda))

plt.bar(y_pos, tempos, align='center', alpha=0.5)
plt.xticks(y_pos, tempos_legenda)
plt.ylabel('Tempo(ms)')
plt.title('Execução dos algoritmos')
plt.figure(figsize=(5,5))
plt.show()
  
```
